```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=100,n_features=6,n_informative=3,n_classes=2)
```

```python
data={f'Feature {i+1}':X[:,i] for i in range(6)}
data['Target']=y
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(3,activation='sigmoid',input_dim=6))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['Target'].values,epochs=100,batch_size=1,verbose=1)
```
